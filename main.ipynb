```python
from transformers import pipeline

corrector = pipeline(
    "text2text-generation",
    model="vennify/t5-base-grammar-correction"
)

def correct(text):
    result = corrector("grammar: " + text, max_new_tokens=100)
    return result[0]["generated_text"]

print(correct("Le vehicule a tombe encontrebas"))
```
