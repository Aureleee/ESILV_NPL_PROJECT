```python
from transformers import pipeline

corrector = pipeline(
    "text2text-generation",
    model="vennify/t5-base-grammar-correction"
)

def correct(text):
    result = corrector("grammar: " + text, max_new_tokens=100)
    return result[0]["generated_text"]

print(correct("Le vehicule a tombe encontrebas"))
```

```python
corrector = pipeline(
    "text2text-generation",
    model="google/flan-t5-xl",
    device_map="auto"
)

def correct(text):
    prompt = f"Correct the grammar and spelling of this French text:\n{text}"
    return corrector(prompt, max_new_tokens=128)[0]["generated_text"]

print(correct("Le vehicule a tombe encontrebas"))
```

```python
from utils_data import (
    get_word_frequencies,
    get_ngram_frequencies,
    prepare_texts_for_lda,
    build_dictionary,
    build_corpus,
    train_lda_model,
    extract_topics
)

import pandas as pd
```

```python
df = pd.read_csv("reviews_concat_prepared.csv")
texts = df["avis_en"].dropna()
```

```python
# n-gram  (not working well)
```

```python

word_freq = get_word_frequencies(texts)

print("Top 20 words:")
print(word_freq.most_common(20))
```

```python
bigram_freq = get_ngram_frequencies(texts, n=2)

print("Top 20 bigrams:")
print(bigram_freq.most_common(20))
```

```python
# LDA 
```

```python
texts_clean = prepare_texts_for_lda(texts)
dictionary = build_dictionary(texts_clean)
corpus = build_corpus(texts_clean, dictionary)
```

```python
lda_model = train_lda_model(
    corpus,
    dictionary,
    num_topics=5,
    passes=10
)
```

```python
topics = extract_topics(lda_model)

for topic_id, topic in topics:
    print(f"Topic {topic_id}: {topic}")
```
